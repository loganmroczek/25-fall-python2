```python
all_permutations = []
def generate_tilings(partial_perm,n):
    global all_permutations

    if sum(partial_perm) == n:
        return partial_perm
    elif sum(partial_perm) > n:
        pass
    else:
        for tile in range(1,5):
            partial_perm.append(tile)
        all_permutations += generate_tilings(partial_perm,n)
    return all_permutations

        
```

```python
print(generate_tilings([],4))
```
