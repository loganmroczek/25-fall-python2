#### Pseudocode

```
value function with the knapsack items as a parameter
    initialize the global variable items and value as 0
    for every item in the knapsack
        increment value by the value of each item
    return value

weight function with the knapsack items as a parameter
    initialize the global variable items and weight as 0
    for every item in the knapsack
        increment weight by the weight of each item
    return weight

knapsack function with capacity as a parameter
    initialize the global variable items 
    initialize knapsack variable with a random index from items
    intitialize most_value variable by calling value function on knapsack

    for each generation in a range from generation 1 to generation 100
        initialize parent variable equal to knapsack 

        for each offspring in range of 100 offspring
            initialize offspring variable as a copy of parent
            initialize offspring_weight varable by calling weight function with the offspring copy of knapsack

            if the weight of offspring is greater than the capacity
                remove a random index in the offspring variable copy of knapsack

            add a random index to the offspring copy of knapsack
            initialize offspring_value variable by calling value function with offspring copy of knapsack
            update offspring_weight variable with new weight after removing and adding items
            
            if offspring_value is greater than most_value and the weight of the offspring knapsack is less than capacity
                set the knapsack to be equal to the offspring
                set the most_value to be equal to the offspring_value
        print the generation number and most_value for each generation
```

````python
def value(knapsack_items):
    '''
    Totals the value of the knapsack items

    :param knapsack_items: list of items in the knapsack
    :return: the total value of the knapsack items
    '''
    global items
    value = 0

    for index in knapsack_items:
        value += items[index][1]

    return value

def weight(knapsack_items):
    '''
    Totals the weigth of the knapsack items

    :param knapsack_items: list of items in the knapsack
    :return: the total weight of the knapsack items
    '''
    global items
    weight = 0

    for index in knapsack_items:
        weight += items[index][0]
    
    return weight

def knapsack(capacity):
    '''
    Evolutionary algorithm that finds the maximum value of the knapsack with weight under the capacity

    :param capacity: the maximum capacity of the knapsack
    '''
    global items
    from random import randint
    
    knapsack = [randint(0,len(items)-1)]
    most_value = value(knapsack)
    parent_value = 0  

    for generation in range(1,1000):
        if most_value == parent_value:
            print(f"Max value reached at {most_value} on generation {generation-2}.")
            break
        
        parent = knapsack
        parent_value = most_value

        for offspring in range(100):
            offspring = parent.copy()
            offspring_weight = weight(offspring)

            if offspring_weight > capacity:
                offspring.pop(randint(0,len(offspring)-1))
            
            offspring.append(randint(0,len(items)-1))

            offspring_value = value(offspring)
            offspring_weight = weight(offspring)

            if offspring_value > most_value and offspring_weight < capacity:
                knapsack = offspring
                most_value = offspring_value
                max_weight = offspring_weight

        print('Generation', generation, ':Value', most_value, ':Weight', max_weight)
````
