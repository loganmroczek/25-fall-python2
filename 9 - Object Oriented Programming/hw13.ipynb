### HW 12: Inheritance

```python
class Point:
    '''
    Represents a point in 2D space

    attributes: x (float), y (float)
    '''
    def __init__(self, input_x = 0, input_y = 0):
        self.x = input_x
        self.y = input_y

    def __str__(self):
        return f'({self.x}, {self.y})'
```

```python
def distance(point1 = Point(), point2 = Point()):
    """
    Returns the distance between two point obkects

    :param point1: point object 1
    :param point2: point object 2
    """
    from math import sqrt

    distance = abs(sqrt((point1.x-point2.x)**2 +(point1.y-point2.y)**2))

    return distance
```

```python
class Polygon:
    '''
    Represents a (possibly degenerate) polygon

    attributes: vertices (list of Point objects)
    '''

    def __init__(self, input_vertices = [Point(), Point(), Point()]):

        self.vertices = input_vertices


    # The __str__ method allows you to specify how an object should be treated when printed.
    def __str__(self):
        
        point_string = 'Vertices: '
        
        # Loop over the list of Points and create a reasonable string representation to return
        for p in self.vertices:
                point_string = point_string + str(p) + ", "

        return point_string[0:-2]
    
    def centroid(self):
        x_total = 0
        y_total = 0
        for point in self.vertices:
            x_total += point.x
            y_total += point.y

        centroid_x = x_total/len(self.vertices)
        centroid_y = y_total/len(self.vertices)

        return Point(centroid_x, centroid_y)
    
    def perimeter(self):
        perimeter = 0
        for index in range(len(self.vertices)):
            perimeter += distance(self.vertices[index-1], self.vertices[index])

        return perimeter
    
    def is_regular(self):
        one_side = distance(self.vertices[-1], self.vertices[0])
        for index in range(1,len(self.vertices)):
            if distance(self.vertices[index-1], self.vertices[index]) != one_side:
                return False
        return True
```

```python
class Triangle(Polygon):
    """
    Represents a (possibly degenerate) triangle

    attributes: vertices (list of Point objects)
    """
    def __init__(self, vertex1 = Point(), vertex2 = Point(), vertex3 = Point()):
        self.vertices = [vertex1, vertex2, vertex3]
```

```python
class Quadrilateral(Polygon):
    """
    Represents a (possibly degenerate) triangle

    attributes: vertices (list of point objects)
    """
    def __init__(self, vertex1 = Point(), vertex2 = Point(), vertex3 = Point(), vertex4 = Point()):
        self.vertices = [vertex1, vertex2, vertex3, vertex4]
```

##### Test Cases

```python
tri = Triangle(Point(0, 0), Point(1, 3), Point(2, 0))
print(tri.centroid())
```

```python
quad = Quadrilateral(Point(0, 0), Point(1, 0), Point(1, 1), Point(0, 1))
print(quad.perimeter())
print(quad.is_regular())
```
