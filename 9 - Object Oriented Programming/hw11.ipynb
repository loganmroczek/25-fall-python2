```python
class Point:
    '''
    Represents a point in 2D space

    attributes: x (float), y (float)
    '''
    def __init__(self, input_x = 0, input_y = 0):
        self.x = input_x
        self.y = input_y

    def __str__(self):
        return f'({self.x}, {self.y})'
```

```python
class Rectangle:
    '''
    Represents a rectangle with lower left corner at corner.

    attributes: corner (Point object), height (float), width (float)
    '''

    # PRACTICE: Define the constructor below
    def __init__(self, input_corner = Point(), input_height = 0, input_width = 0):
        self.corner = input_corner
        self.height = input_height
        self.width = input_width
```

```python
def get_corners(rectObject):
    """
    Returns all corners of rectangle as point objects

    :param rectObject: a rectangle object
    """
    cornerList = [rectObject.corner]
    
    # the 2 x-coordinates and 2 y-cordinates that can be used to get all 4 points
    corner_x = rectObject.corner.x
    corner_y = rectObject.corner.y
    other_x = corner_x + rectObject.width
    other_y = corner_y + rectObject.height

    # add other 3 coordinates to list
    cornerList.append(Point(corner_x, other_y))
    cornerList.append(Point(other_x, other_y))
    cornerList.append(Point(other_x, corner_y))

    return cornerList
```

```python
testRect = Rectangle(Point(0,0), 2, 4)

cornerList = get_corners(testRect)
print(cornerList[0], cornerList[1], cornerList[2], cornerList[3])
```

```python
def distance(point1 = Point(), point2 = Point()):
    """
    Returns the distance between two point obkects

    :param point1: point object 1
    :param point2: point object 2
    """
    from math import sqrt

    distance = sqrt((point1.x-point2.x)**2 +(point1.y-point2.y)**2)

    return distance
```

```python
print(distance(Point(), Point(1,1)))
```

```python
class Circle:
    """
    Circle object represented by its center point and the radius of the circle

    attributes: center(point), radius(float)
    """
    def __init__(self, center = Point(), radius = 0):
        self.center = center
        self.radius = radius
```

```python
def point_in_circle(circleObject, pointObject):
    """
    Checks if the point is within the circle

    :param cicleObject: circle object
    :param pointObject: point object
    """
    distance_from_center = abs(distance(pointObject, circleObject.center))

    if distance_from_center <= circleObject.radius:
        return True
    else:
        return False
```

```python
circleTest = Circle(Point(), 1)

print(point_in_circle(circleTest, Point(1,0)))
print(point_in_circle(circleTest, Point(1.1,0)))
```

```python
def rect_in_circle(circleObj, rectObj):
    """
    Checks if all the points of a rectangle are within the cicle

    :param circleObj: circle object
    :param rectObj: rectangle object
    """
    rectCorners = get_corners(rectObj)

    for point in rectCorners:
        if point_in_circle(circleObj, point) == False:
            return False
    
    return True
```
