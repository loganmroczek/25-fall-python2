```python
class Point:
    '''
    Represents a point in 2D space

    attributes: x (float), y (float)
    '''
    def __init__(self, input_x = 0, input_y = 0):
        self.x = input_x
        self.y = input_y

    def __str__(self):
        return f'({self.x}, {self.y})'
```

```python
class Circle:
    """
    Circle object represented by its center point and the radius of the circle

    attributes: center(point), radius(float)
    """
    def __init__(self, center = Point(), radius = 0):
        self.center = center
        self.radius = radius

    def __str__(self):
        return f'Center: {self.center}, Radius: {self.radius}'

    def get_area(self):
        from math import pi
        area = pi * self.radius**2
        return area
    
    def to_square(self):
        from math import sqrt
        side = sqrt(self.get_area())
        corner_x = self.center.x - side/2
        corner_y = self.center.y - side/2

        return Rectangle(Point(corner_x, corner_y), side, side)
```

```python
class Rectangle:
    '''
    Represents a rectangle with lower left corner at corner.

    attributes: corner (Point object), height (float), width (float)
    '''

    # PRACTICE: Define the constructor below
    def __init__(self, input_corner = Point(), input_height = 0, input_width = 0):
        self.corner = input_corner
        self.height = input_height
        self.width = input_width

    def __str__(self):
        return f'Corner: {self.corner}, Height: {self.height}, Width: {self.width}'

    def get_area(self):
        area = self.height * self.width
        return area

    def __lt__(self, other):
        if self.get_area() < other.get_area(): 
            return True
        else:
            return False
        
    def get_center_point(self):
        input_x = self.corner.x
        input_y = self.corner.y
        input_height = self.height
        input_width = self.width

        midpoint_x = input_x+(input_width/2)
        midpoint_y = input_y+(input_height/2)
        
        return Point(midpoint_x, midpoint_y)

    def to_circle(self):
        from math import sqrt, pi
        radius = sqrt(self.get_area()/pi)
        return Circle(self.get_center_point(), radius)


```

```python
rect = Rectangle(Point(1,2), 3, 4)
print(rect.get_area())
```

```python
rect2 = Rectangle(Point(), 6, 6)
print(rect < rect2)
```

```python
print(rect.to_circle())
```
